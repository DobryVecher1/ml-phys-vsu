# Примеры использования библиотеки Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-15, 15, 100)

plt.plot(np.sin(t)/t, '-')
plt.plot(t, np.cos(t), '--')
```

```python
# Matlab style
plt.figure()  # Создаем фигуру

# Рисуем график на первом subplot
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(t, np.sin(t))

# На втором
plt.subplot(2, 1, 2)
plt.plot(t, np.cos(t))
```

```python
# Объектно-ориентированный подход
fig, ax = plt.subplots(2)
# Вызываем plot к соответствующему объекту
ax[0].plot(t, np.sin(t))
ax[1].plot(t, np.cos(t))
```

```python
fig1, ax1 = plt.subplots(3, 2, figsize=(10, 6))
```

```python
fig = plt.figure()
ax = plt.axes()
ax.grid()
```

```python
plt.plot(t, np.sin(t - 0), color='blue')        # specify color by name
plt.plot(t, np.sin(t - 1), color='g')           # short color code (rgbcmyk)
plt.plot(t, np.sin(t - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(t, np.sin(t - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(t, np.sin(t - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(t, np.sin(t - 5), color='chartreuse'); # all HTML color names supported
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# Более короткая запись
plt.plot(x, x + 4, linestyle='-')
plt.plot(x, x + 5, linestyle='--')
plt.plot(x, x + 6, linestyle='-.')
plt.plot(x, x + 7, linestyle=':');
```

```python
# Пределы графика
plt.plot(x, np.sin(x))

plt.xlim(-5, 11)
plt.ylim(-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]) # Формат [xmin, xmax, ymin, ymax]
```

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')

plt.legend()
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot')
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.plot(x, y, '*', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2)
```

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
plt.contour(X, Y, Z, colors='black')
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
```
